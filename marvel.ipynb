```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
import re 
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films"
headers = {'User-Agent': 'UniProjectSimple/1.0 (karthik018hv@gmail.com)'}

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.content, 'html.parser')

table = soup.find_all('table',class_='wikitable')

print(table)
```

```python
# The film data is typically in the FIRST 'wikitable' found, 
# which is the comprehensive list of films.
# We convert the selected table element to a string before passing to read_html
html_string = str(table[0]) 

# Use pd.read_html to parse the HTML string directly into a list of DataFrames.
# Since we only passed one table string, the list will contain one DataFrame.
df_list = pd.read_html(html_string, header=0)

# Extract the single DataFrame from the list
df = df_list[0]

# Print the first few rows and column info for verification
print("--- DataFrame Head ---")
print(df.head().to_markdown(index=False))
print("\n--- DataFrame Info ---")
df.info()
```
